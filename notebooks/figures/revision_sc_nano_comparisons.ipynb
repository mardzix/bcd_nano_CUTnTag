```r
library(cluster, quietly = TRUE)

get_silhouette_score <- function(dataset, celltype_var, reduction='lsi', dims=2:40){
    dist.matrix <- dist(x = Embeddings(object = dataset[[reduction]])[, dims])
    clusters <- dataset@meta.data[,celltype_var]
    sil <- silhouette(x = as.numeric(x = as.factor(x = clusters)), dist = dist.matrix)
    return(sil)
    }

sil.nbiotech <- get_silhouette_score(dataset = seurat.nbiotech,
                                     celltype_var = 'cell_type')

sil.nano <- get_silhouette_score(dataset = seurat.nano,
                                 celltype_var = 'idents_short')
```

```r
seurat.nbiotech <- AddMetaData(seurat.nbiotech,metadata = sil.nbiotech[,3],col.name = 'silhouette_score')
seurat.nano     <- AddMetaData(seurat.nano,metadata = sil.nano[,3],col.name = 'silhouette_score')

FeaturePlot(seurat.nbiotech,'silhouette_score') + scale_color_viridis_c()
FeaturePlot(seurat.nano,'silhouette_score') + scale_color_viridis_c()
```

```r
df1 <- data.frame('silhouette_score' = seurat.nbiotech$silhouette_score,'dataset'='nbiotech')
df2 <- data.frame('silhouette_score' = seurat.nano$silhouette_score,'dataset'='nano-CT')

df.to.plot <- rbind(df1,df2)
ggplot(data=df.to.plot) + geom_violin(aes(x=dataset,y=silhouette_score, fill=dataset))
```
